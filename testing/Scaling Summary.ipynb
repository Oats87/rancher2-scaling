```python
from matplotlib import pyplot
import pandas

data = pandas.read_csv('scale_test.csv')
pyplot.plot(data["k8s_cluster_list_time"], label="k8s")
pyplot.plot(data["rancher_cluster_list_time"], label="rancher")
pyplot.xlabel("iteration")
pyplot.ylabel("seconds")
pyplot.title("cluster list times")
pyplot.legend()
pyplot.savefig('graph.png')

```
